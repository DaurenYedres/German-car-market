T - Test

```python
from scipy.stats import ttest_ind
import scipy.stats as sc
```

The result of t- test is the P value. If P value > 0.05, It means there is no significant difference. But if P value < 0.05 then it means there IS a significant difference.

P value is the probability of the corectness of the NULL hypothesis which means that both groups have the same scores.

```python
df_bmw.describe()
```

```python
ttest_ind(df_bmw['mileage'], df_bmw['price'])
```

T - test shows that all numeric groups of the dataset have significance difference between each other.

There is no sense to compare lists of price and mileage for example, because their values are different.
